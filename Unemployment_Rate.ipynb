# Introduction


**Unemployment Rate and Its Importance**:
- Unemployment is a critical economic indicator that reflects the health of an economy. The unemployment rate is the percentage of people who are unemployed out of the total labor force. A high unemployment rate indicates economic distress, while a low rate suggests economic stability and growth.

**Objective of the Analysis**:
- This analysis aims to explore and understand the trends in the unemployment rate across various states in India. Special attention is given to the impact of Covid-19 on unemployment rates, providing insights into how different regions were affected.


# Data Description


**Overview of the Dataset:**
- The dataset contains the unemployment rate of all the states in India, observed monthly. It covers a period that includes significant economic events, such as the Covid-19 pandemic.

**Key Columns in the Dataset:**
- Region: States in India.
- Date: The date when the unemployment rate was observed.
- Frequency: The frequency of measurement (Monthly).
- Estimated Unemployment Rate (%): The percentage of people unemployed in each state.
- Estimated Employed: The percentage of people employed.
- Estimated Labour Participation Rate (%): The percentage of people actively participating in the labor force.


# Data Cleaning


**Addressing Missing Values:**
- Missing values were identified and handled to ensure the dataset's integrity. Rows with missing values were removed to maintain the consistency and accuracy of the analysis.

**Removing Outliers:**
- Outliers can skew the analysis and lead to misleading conclusions. Using the Interquartile Range (IQR) method, outliers were detected and removed from the dataset. This ensures a more reliable analysis.

# Regional Comparison


### Average Unemployment Rates Across States


A bar plot was used to compare the average unemployment rate across different states.


**Key Insight:**

Some regions have consistently higher unemployment rates than others, indicating regional disparities that may be due to various economic and social factors.

#### distribution of unemployment rates across different regions.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df1)
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Distribution by Region')
plt.xticks(rotation=90)
plt.show()
```

Here are the summarized insights from the graph:

1. **High Unemployment Regions:**
   - Haryana, Himachal Pradesh, and Delhi have the highest median unemployment rates with significant variability.

2. **Low Unemployment Regions:**
   - Andhra Pradesh, Gujarat, Maharashtra, and Uttar Pradesh show lower median unemployment rates with more consistent distribution.

3. **Outliers:**
   - Regions like Goa, Jammu & Kashmir, and Telangana exhibit significant outliers, indicating periods or areas with extremely high unemployment rates.

4. **Regional Patterns:**
   - Northern and Western states generally have higher unemployment rates compared to Southern states.
   - Eastern states like Bihar and Assam show moderate unemployment rates with considerable variability.

5. **Overall Variability:**
   - The unemployment rates during COVID-19 varied widely across regions, highlighting the uneven economic impact of the pandemic.

#### Heat Maps

# Heatmap Analysis


### Visual Representation of Unemployment Trends


A heatmap was created to visualize the unemployment rate over time for each region. This helps identify periods of high and low unemployment rates.


**Key Insight:**

Some regions experienced more drastic changes in unemployment rates compared to others. This visual representation helps pinpoint when and where significant changes occurred.

```python
# Pivot data for heatmap
pivot_table = df1.pivot_table(values=' Estimated Unemployment Rate (%)', index=' Date', columns='Region')

# Heatmap for Unemployment Rate by Region
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap="YlGnBu", cbar_kws={'label': 'Unemployment Rate (%)'})
plt.xlabel('Region')
plt.ylabel('Date')
plt.title('Heatmap of Unemployment Rate Over Time by Region')
plt.show()
```

```python
# Scatter plot of unemployment rate over time
df1['Date'] = pd.to_datetime(df1['Date'])  # Convert to datetime
plt.figure(figsize=(10, 6))
plt.scatter(df1['Date'], df1[' Estimated Unemployment Rate (%)'])
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.show()
```


        
**The unemployment rate remains relatively stable from mid-2019 to early 2020.
There is a significant increase in the unemployment rate starting around March 2020, reaching peaks in the subsequent months.
This increase around early 2020 likely correlates with global events such as the COVID-19 pandemic, which led to widespread economic disruptions and increased unemployment rates worldwide.**



```python
# Correlation between unemployment rate and other numerical features
(df1[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr())
```

```python
# Scatter plot of unemployment rate vs. labor participation rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x=' Estimated Labour Participation Rate (%)', y=' Estimated Unemployment Rate (%)', data=df1)
plt.title('Unemployment Rate vs. Labour Participation Rate')
plt.xlabel('Estimated Labour Participation Rate (%)')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

**From the scatter plot, the following observations can be made:**

- The data points for the unemployment rate and labor force participation rate are quite dispersed, but some trends can be seen.
- Higher labor force participation rates (above 50%) tend to correspond with lower unemployment rates.
- Lower labor force participation rates (below 40%) correspond with a wide range of unemployment rates, from 0% to 70%.
- This chart can be used to analyze the relationship between the unemployment rate and the labor force participation rate, helping to identify potential economic patterns and trends.
